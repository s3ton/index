```python
from abc import ABC, abstractmethod
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.graph_objects as go  
import json
from classi_strategia import VolMktCapStrategy, RebalancedMarketCapStrategy, MarketCapThresholdStrategy
```

```python
# ticker
with open("ticker_yf.json", "r") as f:
    dati_caricati = json.load(f)

conio_coin = dati_caricati['conio_coin']
tech = dati_caricati['tech']
```

```python
# 1. Inizializziamo la strategia figlia con un capitale di partenza
strategy = VolMktCapStrategy(initial_value=10000.0 )

# 2. Importiamo i dati
strategy.import_data(assets=conio_coin, timeframe="5y")

# 3. Eseguiamo il backtest (es. ribilanciamento ogni 30 giorni)

strategy.run_strategy(rebalance_freq_days=180)

# 4. Calcoliamo e stampiamo le statistiche
stats_df = strategy.print_stats()

# 5. Mostriamo il grafico
strategy.plot_results()
```
