```python
import json
import os
import sys
from data_loader import download_all_market_data
from strategies import RebalancedMarketCapStrategy, MarketCapThresholdStrategy, GlobalMarketAnalytics

# Soluzione per script (.py) o notebook (.ipynb)
try:
    SCRIPT_DIR = os.path.dirname(os.path.abspath(__file__))
except NameError:
    # Se __file__ non esiste, siamo in un notebook
    SCRIPT_DIR = os.path.abspath('') 

DATA_DIR = os.path.join(SCRIPT_DIR, "data")
CHARTS_BASE_DIR = os.path.join(SCRIPT_DIR, "../charts")


# 1. Caricamento categorie da portfolios.json
try:
    with open(os.path.join(SCRIPT_DIR, "portfolios.json"), "r") as f:
        portfolios = json.load(f)
except FileNotFoundError:
    print("Errore: file 'portfolios.json' non trovato.")
    sys.exit(1)

conio_coin = portfolios.get("conio_coin", [])
tech = portfolios.get("tech", [])
defi = portfolios.get("defi", [])
pagamenti = portfolios.get("pagamenti", [])

# ==========================================
# 2. DOWNLOAD E AGGIORNAMENTO DATI
# ==========================================
print("\n--- Download Dati Centralizzato (CSV + API) ---")
# Uniamo tutti gli asset di tutti i portafogli per un aggiornamento completo
all_unique_assets = list(set(conio_coin + tech + defi + pagamenti))

download_all_market_data(
    assets=all_unique_assets, 
    timeframe="5y",          
    data_dir=DATA_DIR,          
    benchmark_ticker="BTC"    
)

# ==========================================
# 3. ANALISI E BACKTEST
# ==========================================

# --- Top 5 Market Cap ---
print("\n--- Analisi: Top 5 Market Cap ---")
strategy_5mktcap = RebalancedMarketCapStrategy()
strategy_5mktcap.load_local_data(assets=conio_coin, data_dir=DATA_DIR)
strategy_5mktcap.run_strategy()
strategy_5mktcap.plot_results(
    title="Top 5 Market Cap Index vs BTC", 
    save_json_dir=os.path.join(CHARTS_BASE_DIR, "Top5MktCap")
)
# Nota: salviamo anche l'evoluzione dei pesi
strategy_5mktcap.plot_weights(save_json_dir=os.path.join(CHARTS_BASE_DIR, "Top5MktCap"))

# --- Settore Tech ---
print("\n--- Analisi: Tech Crypto Market ---")
strategy_threshold_tech = MarketCapThresholdStrategy()
strategy_threshold_tech.load_local_data(assets=tech, data_dir=DATA_DIR)
strategy_threshold_tech.run_strategy()
strategy_threshold_tech.plot_results(
    title="Tech crypto market index", 
    save_json_dir=os.path.join(CHARTS_BASE_DIR, "TechMktCap_TH")
)

# --- Settore DeFi ---
print("\n--- Analisi: DeFi Crypto Market ---")
strategy_threshold_defi = MarketCapThresholdStrategy()
strategy_threshold_defi.load_local_data(assets=defi, data_dir=DATA_DIR)
strategy_threshold_defi.run_strategy()
strategy_threshold_defi.plot_results(
    title="Defi crypto market index", 
    save_json_dir=os.path.join(CHARTS_BASE_DIR, "DefiMktCap_TH")
)

# --- Settore Pagamenti ---
print("\n--- Analisi: Pagamenti Crypto Market ---")
strategy_threshold_pag = MarketCapThresholdStrategy()
strategy_threshold_pag.load_local_data(assets=pagamenti, data_dir=DATA_DIR)
strategy_threshold_pag.run_strategy()
strategy_threshold_pag.plot_results(
    title="Pagamenti crypto market index", 
    save_json_dir=os.path.join(CHARTS_BASE_DIR, "PagamentiMktCap_TH")
)

# ==========================================
# 4. MARKET ANALYTICS
# ==========================================
print("\n--- Generazione Market Analytics ---")
market_analytics = GlobalMarketAnalytics()
market_analytics.load_local_data(assets=conio_coin, data_dir=DATA_DIR)
market_analytics.generate_analytics_charts(
    conio_coin=conio_coin, 
    save_dir=os.path.join(CHARTS_BASE_DIR, "market_analytics"),
    tf_dom='5y',       
    tf_vol='1y',  
    show_figs=True
)

print("\n!!! ANALISI COMPLETATA !!!")
```
