```python
from abc import ABC, abstractmethod
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.graph_objects as go  
import json
from classi_strategia import VolMktCapStrategy, RebalancedMarketCapStrategy, MarketCapThresholdStrategy, EqualWeightThresholdStrategy, GlobalMarketAnalytics

# diagnostics: show which Python/kernel is running (useful when kernel errors occur)
import sys, os
print("Python executable:", sys.executable)
print("Python version:", sys.version)
print("Notebook cwd:", os.getcwd())
```

```python
with open("ticker_yf.json", "r") as f:
    dati_caricati = json.load(f)

conio_coin = dati_caricati['conio_coin']
tech = dati_caricati['tech']
defi = dati_caricati['defi']
pagamenti = dati_caricati['pagamenti']
```

```python
# marketcap
strategy_5mktcap = RebalancedMarketCapStrategy()
strategy_5mktcap.import_data(assets= conio_coin, timeframe="5y")
strategy_5mktcap.run_strategy()
strategy_5mktcap.plot_results(title="Top 5 Market Cap Index vs BTC-USD",
                              save_json_dir="../charts/Top5MktCap")
strategy_5mktcap.plot_weights()
#tech
strategy_threshold = MarketCapThresholdStrategy()
strategy_threshold.import_data(assets=tech, timeframe="5y")
strategy_threshold.run_strategy()
strategy_threshold.plot_results(title="Tech crypto market index",
                                save_json_dir="../charts/TechMktCap_TH")
#defi
strategy_threshold = MarketCapThresholdStrategy()
strategy_threshold.import_data(assets=defi, timeframe="5y")
strategy_threshold.run_strategy()
strategy_threshold.plot_results(title="Defi crypto market index",
                                save_json_dir="../charts/DefiMktCap_TH")
#pagamenti
strategy_threshold = MarketCapThresholdStrategy()
strategy_threshold.import_data(assets=pagamenti, timeframe="5y")
strategy_threshold.run_strategy()
strategy_threshold.plot_results(title="Pagamenti crypto market index",
                                save_json_dir="../charts/PagamentiMktCap_TH")
#market analytics
market_analytics = GlobalMarketAnalytics()
market_analytics.import_data(conio_coin, timeframe="5y")
market_analytics.generate_analytics_charts(
    conio_coin=conio_coin, 
    save_dir="../charts/market_analytics",
    tf_dom='5y',       # Dominance dell'ultimo anno
    tf_vol='1y',  # Correlazione calcolata sugli ultimi 6 mesi
    show_figs=True)
```
